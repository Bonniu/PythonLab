## Część na ocenę 3

#### Importowanie danych do zadania na ocenę 3

```python
dane = pd.read_csv("fertility_Diagnosis.data", names=['Season', 'Age', 'Dis', 'Trau', 'Inter', 'Fever',
                             'FreqAlc', 'SmokingH', 'SitHours', 'Output'])
ilosc_danych: int = len(dane)
```

#### Cecha jakościowa Season 

```python
print("\nCecha jakościowa: Season")
season = dane.Season
dominant = statistics.mode(season)
licznik = len([1 for i in season if i == dominant])
slownik: dict = {-1: 'zima', -0.33: 'wiosna', 0.33: 'lato', 1: 'jesień'}
print("Dominanta: ", slownik[dominant])
print("Liczebność: ", licznik)
print("Częstość: ", licznik / ilosc_danych)
```

#### Cecha ilościowa Age

```python
print("\nCecha ilościowa: Age")
print("Opis: Wiek w momencie analizy. 18–36 (0, 1)")
age = round(dane.Age * 18) + 18
print("Średnia: ", np.mean(age))
print("Odchylenie standardowe: ", np.std(age))
print("Mediana: ", np.median(age))
print("Maksimum: ", np.amax(age))
print("Minimum: ", np.amin(age))
```

#### Krzywa regresji dla dwóch cech ilościowych najbardziej ze sobą skorelowanych

```python
# przypisujemy dane do odpowiednich zmiennych
x = dane_2['petal_width'].values.reshape(-1, 1)
y = dane_2['petal_length']

# Dzielimy dane - 50% danych jako zbiór testowy, 50% jak zbiór treningowy
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.50)

# Utworzenie nowego obiektu klasy LinearRegression
regressor = LinearRegression()
# Trenowanie z użyciem danych x_train i y_train
regressor.fit(x, y)
# Utworzenie przewidywanego y
y_pred = regressor.predict(x)

# rysowanie wykresu
plt.figure(figsize=[12,12])
plt.scatter(x, y,  color='red')
plt.plot(x, y_pred, color='violet', linewidth=4)
plt.gca().set(title='Krzywa regresji cech ilościowych (petal_length i petal_width)', 
              ylabel='petal_length', 
              xlabel='petal_width')

plt.show()
```
